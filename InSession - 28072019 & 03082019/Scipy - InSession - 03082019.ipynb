Scores on an IQ test have a mean 100 and a standard deviation of 15. Find ℙ(85 < 𝑋 < 115)

X follows N(100,15)

(85-100)/15 = -1 & (115-85)/100 = 1

P(85 < X < 115) = P(-1 < X < 1)

= P(x < 1) - P( x < -1)


```python
# P(x < 1) - P( x < -1)

norm.cdf(1) - norm.cdf(-1)

# norm.cdf -- area under the curve towards the left for any point
```

```python
norm.cdf(115, loc = 100, scale = 15) - norm.cdf(85, 100, 15)

# default mean = 0 and default sd = 1
# cdf cumulative distribution function
# loc means mean and scale means standard deviation
# default values of loc and scale are 0 and 1 respectively
```

```python
print (norm.cdf(1) - norm.cdf(-1)) # area b/w first std
print (norm.cdf(2) - norm.cdf(-2)) # area b/w second std
print (norm.cdf(3) - norm.cdf(-3)) # area b/w third std
```

1.
The mean lifetime of a tire is 30,000 km. The standard deviation is 2000 km. 
If a company purchased 2000 tires, assuming a normal distribution, how many 
tires would last more than 28,000 kms?

Ans:
X = mean lifetime of a tire;   X follows N(30,000, 2000);   n = 2000

Let us check the prob of 1 tire lasting more than 28000 kms

P(x > 28000) = 1 - P(x < 28000)

Standardize 28000 --- (28000-30000)/2000 = -1

P(x > 28000) = 1 - P(x < 28000) = 1 - P( z < -1)

```python
# 1 - P( z < -1)

prob = 1 - norm.cdf(-1) # prob of 1 tire lasts more than 28000 kms

print (prob)
2000*prob
```

```python
# P(x > 28000) = 1 - P(x < 28000)

prob = 1 - norm.cdf(28000, 30000, 2000) # prob of 1 tire lasts more than 28000 kms

print (prob)

2000*prob
```
